```python
Outcome_Vals_Cat = {
    1: "diabetic",
    0: "non_diabetic",
}

Outcome_Vals_Num = {
    "diabetic": 1,
    "non_diabetic": 0,
}

df["Outcome"] = df["Outcome"].replace(Outcome_Vals_Cat)
df
```

```python
Outcome_counts = df['Outcome'].value_counts().reindex(['diabetic', 'non_diabetic'])

colors = ['#bcbd22', '#17becf']

bar_width = 0.8

plt.bar(Outcome_counts.index, Outcome_counts.values, width=bar_width, color=colors)

plt.title('Number of Samples by Outcome')

plt.xlabel('Outcome')
plt.ylabel('Number of Samples')

for i, count in enumerate(Outcome_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.show()
```
